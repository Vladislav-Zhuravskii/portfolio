## Описание задачи

Оператор связи хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

### Описание услуг

Оператор предоставляет два основных типа услуг: 

- Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.

- Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

### Описание данных

Данные состоят из файлов, полученных из разных источников:

- `contract.csv` — информация о договоре;
   - `BeginDate` – дата начала пользования услугами,
   - `EndDate` – дата окончания пользования услугами,
   - `Type` – тип оплаты: ежемесячный, годовой и тд,
   - `PaperlessBilling` – безналичный расчет,
   - `PaymentMethod` – способ оплаты,
   - `MonthlyCharges` – ежемесячные траты на услуги,
   - `TotalCharges` – всего потрачено денег на услуги
   
   
- `personal.csv` — персональные данные клиента;
   - `gender` - пол клиента
   - `Dependents` – наличие иждивенцев
   - `Senior Citizen` – наличие пенсионного статуса по возрасту
   - `Partner` – наличие супруга(и)
   
   
- `internet.csv` — информация об интернет-услугах;


- `phone.csv` — информация об услугах телефонии.
   - `MultipleLines` – наличие возможности ведения параллельных линий во время звонка

Во всех файлах столбец `customerID` содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.

## Предобработка данных

### Пропуски

Заполним появившиеся после слияния таблиц пропуски. Для признаков, связанных с инернет-обслуживанием:

* Заполним значением `'No Internet'` ("нет интернет-обслуживания") - признак `InternetService`.

* Заполним значением `'No'` - дополнительные интернет-услуги.

````python
data['InternetService'] = data['InternetService'].fillna(value='No Internet')
internet_addons = [
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies'
]
data[internet_addons] = data[internet_addons].fillna(value='No')
````

Преобразуем признак `MultipleLines` (наличие нескольких телефонных линий) в новый признак `LandlineType` (тип телефонной линии с категорией `'No line'` (нет телефонного обслуживания) для пропусков.

````python
def landline_type(value):
    if value == 'Yes':
        return 'Multiple lines'
    elif value == 'No':
        return 'Single line'
    else:        
        return 'No line'

data['LandlineType'] = data['MultipleLines'].apply(landline_type)

data = data.drop('MultipleLines', axis=1)
````

Проверим, все ли пропуски заполнены.

````python
print('Количество пропусков в данных после заполнения:', data.isna().sum().sum())
````

### Целевой признак

Выделим целевой признак из столбца `EndDate`. Для начала посмотрим уникальные значения для этого признака.

````python
data['EndDate'].value_counts()
````

Наблюдаем, что признак принимает либо значение `'No'` - очевидно, для клиентов, не прервавших договор с компанией, либо значение даты разрыва договора. Обратим внимание, что даты принимают значения только в пределах последних четырех месяцев (от даты актуальности данных - 1 февраля 2020). Таким образом можно сделать вывод, что данные были собраны для пользователей, ушедших недавно (за последние четыре месяца), либо остающихся клиентами. Такая выборка хорошо подойдет для бинарной классификации: данные о пользователях, ушедших годы назад, скорее всего слабо помогли бы в построении модели, предсказывающей, собирается ли пользователь уйти в скором времени.

Создадим целевой признак `client_left` (ушел ли клиент недавно) со значениями:

* «1» - для ушедших (есть дата окончания договора).

* «0» - для оставшихся (договор не окончен).

````python
data['client_left'] = data['EndDate'].apply(lambda x: 0 if x == 'No' else 1)
````

### Длительность договора

Преобразуем признаки даты накала и окончания договора с клиентом. Сами по себе для решения задачи классификации они не очень информативны. Однако из них можно создать новый признак - длительность действия договора (до его окончания или на текуший момент). Для начала запишем в константу значение текушей даты (даты актуальности данных).

````python
ACTUAL_DATE = '2020-02-01'
````

Преобразуем `BeginDate` в формат `datetime`.

````python
data['BeginDate'] = pd.to_datetime(data['BeginDate'], format='%Y-%m-%d')
````

Преобразуем `EndDate` в формат `datetime`, заменив для оставшихся пользователей значения на актуальную дату.

````python
data['EndDate'] = pd.to_datetime(
    data['EndDate'].where(data['EndDate'] != 'No', ACTUAL_DATE),
    format='%Y-%m-%d'
)    
````

Периодичность данных в этих признаках - месяцы. Создадим признак `elapsed_months` (длительность действия договора в месяцах).

````python
data['elapsed_months'] = (
    (data['EndDate'].dt.year - data['BeginDate'].dt.year) * 12 + 
    (data['EndDate'].dt.month - data['BeginDate'].dt.month)
)
````

Посмотрим, есть ли в данных клиенты, заключившие тоговор долько что (срок обслуживания 0 месяцев).

````python
data[data['elapsed_months'] == 0]
````

Да, есть 11 таких клиентов. Заметим также, что для них нет значений `TotalCharges` - они еще не успели оплатить услуги телеком-компании. Данные об этих клиентах не помогут в построении модели - на момент актуальности данных они еще не успели воспользоваться услугами компании и принять решение, уйти им или оставаться. Отбросим эти значения.

````python
data_len = len(data)
data = data.drop(index=data[data['elapsed_months'] == 0].index)
print('Отброшено {:.3%} данных'.format((data_len - len(data))/data_len))
````

Отбросим данные с датами: столбец `EndDate` содержит информацию о целевом признаке, что вызовет утечку данных, а `BeginDate` не особо нужен, так как мы создали признак длительности обслуживания клиента компанией.

````python
data = data.drop(['BeginDate', 'EndDate'], axis=1)
````

### Типы данных

````python
data.info()
````

Изменим тип данных для `TotalCharges` на численный.

````python
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'])
````

В данных много категориальных признаков, принимающих бинарные знанчения (Yes / No). Их можно поменять на 1 / 0 - так уже закодирован, например, признак `SeniorCitizen`.

````python
binary_columns = [
    col for col in data.columns 
    if sorted(data[col].unique().tolist()) == ['No', 'Yes']
]

for column in binary_columns:
    data[column] = data[column].apply(lambda x: 1 if x == 'Yes' else 0)
````

Добавим также признак `internet_addons` - сколько всего дополнительных услуг, связанных с интернетом, подключено у клиента.

````python
data['internet_addons'] = data[internet_addons].sum(axis=1)
````

Для экономии памяти поменяем тип целочисленных признаков на `int8`.

Исследуем признаки, связанные с оплатой. Для удобства построения однотипных графиков напишем функци:

````python
def add_barplot(data, feature, row, col, percentage=False):
    d = {1: 'Да', 0: 'Нет'}
    clients_left = data.groupby(feature).sum()['client_left'].sort_index()
    total_clients = data.groupby(feature).count()['client_left'].sort_index()
    
    if percentage == False:
        clients_stayed = total_clients - clients_left    

        fig.add_trace(go.Bar(
            x=[d[x] if x in d else x for x in sorted(data[feature].unique())],
            y=clients_stayed,
            name='Клиент остался',
            marker_color='LimeGreen',
            hovertemplate =
                'Количество клиетов: %{y}'+
                '<br>%{x}',
        ), row=row, col=col)

        fig.add_trace(go.Bar(
            x=[d[x] if x in d else x for x in sorted(data[feature].unique())],
            y=clients_left,
            name='Клиет ушёл',
            marker_color='OrangeRed',
            hovertemplate =
                'Количество клиетов: %{y}'+
                '<br>%{x}',
        ), row=row, col=col)
        
    else:
        left_percentage = clients_left / total_clients * 100
        
        fig.add_trace(go.Bar(
            x=[d[x] if x in d else x for x in sorted(data[feature].unique())],
            y=left_percentage,
            text=['{:.1f}%'.format(x) for x in left_percentage],
            name='Процент ушедших',
            hovertemplate =
                'Процент ушедших: %{y:.2f}%'+
                '<br>%{x}',
            marker={'color': left_percentage,
                    'colorscale': 'Viridis'}
        ), row=row, col=col)
````

### Проверка гипотез

Проверим наличие связи между признаками и целевым признаком с помощью методов статистики `scipy.stats`.

Для проверки связи количественных признаков `TotalCharges`, `MonthlyCharges`, `elapsed_months` и номинального целевого признака можно сформулировать гипотезы о равенстве средних значения признака для ушедших и оставшихся клиентов и проверить их.

Для кажного признака сформулируем нулевую и альтернативную гипотезы:

* H0: средние значения признака для ушедших и оставшихся клиентов равны.
* H1: средние значения признака для ушедших и оставшихся клиентов различаются.

Зададим значение статистической значимости в 1%.

````python
significance_level = 0.01
````

В цикле для каждого количественного признака:

* Разделим выборки признака для ушедших и оставшихся, рассчитаем средние значения.

* Проверим равенство дисперсий выборок. Рассчитаем критерий Бартлетта с помощью `scipy.stats.bartlett`. Нулевая гипотеза этого теста - дисперсии равны.

* Проверим гипотезу с помощью `scipy.stats.ttest_ind`, т.к. выборки не являются зависимыми. Укажем аргумент для метода `equal_var` - равенство дисперсий. Метод будет считать t-критерий Стьюдента в случае равенства дисперсий или t-критерий Уэлча - в случае неравенства.

Соберем результаты в таблицу.

````python
features = ['TotalCharges', 'MonthlyCharges', 'elapsed_months']

result = pd.DataFrame()
for feature in features:
    left = data.query('client_left == 1')[feature]
    stayed = data.query('client_left == 0')[feature]
    result.loc[feature, 'Среднее ушедших'] = round(left.mean(), 2)
    result.loc[feature, 'Среднее оставшихся'] = round(stayed.mean(), 2)
    
    test_results = st.bartlett(left, stayed)
    if test_results.pvalue <= significance_level:
        result.loc[feature, 'Дисперсии'] = 'не равны'
        equal_var=False
    else:
        result.loc[feature, 'Дисперсии'] = 'равны'
        equal_var=True
    
    test_results = st.ttest_ind(left, stayed, equal_var=equal_var)
    result.loc[feature, 'P-значение'] = test_results.pvalue
    if test_results.pvalue <= significance_level: 
        result.loc[feature, 'Принятая гипотеза'] = 'H1: средние значения различаются'
    else: 
        result.loc[feature, 'Принятая гипотеза'] = 'H0: средние значения равны'
    
display(result)
````

Для всез признаков нулевая гипотеза H0 отвергнута и принята альтернативная гипотеза H1: средние значения признака для ушедших и оставшихся клиентов различаются.

Для проверки связи всех остальных признаков (номинальных) и номинального целевого признака можно сформулировать гипотезы о зависимости признаков и рассчитать критерий Хи-квадрат.

Для кажного признака сформулируем нулевую и альтернативную гипотезы:

* H0: признак и целевой признак независимы.
* H1: признак и целевой признак не являются независимыми.

В цикле по всем оставшимся признакам проверим гипотезы с помощью `scipy.stats.chi2_contingency`. Выведем в форме таблицы степени свободы, P-значение и принятую гипотезу для каждого признака.

````python
features = [
    'Type', 'PaperlessBilling', 'PaymentMethod',
    'InternetService', 'internet_addons', 'OnlineSecurity', 'OnlineBackup',
    'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
    'gender', 'SeniorCitizen', 'Partner', 'Dependents', 'LandlineType',
]

result = pd.DataFrame()
for feature in features:
    chisqt = pd.crosstab(data['client_left'], data[feature], margins=True)
    value = np.array([chisqt.iloc[0].values,
                      chisqt.iloc[1].values])
    p, dof = st.chi2_contingency(value)[1:3]
    result.loc[feature, 'Степени свободы'] = dof
    result.loc[feature, 'P-значение'] = p
    if p <= significance_level: 
        result.loc[feature, 'Принятая гипотеза'] = 'H1: признаки не являются независимыми'
    else: 
        result.loc[feature, 'Принятая гипотеза'] = 'H0: признаки независимы'
        
    result = result.sort_values(by='P-значение')

display(result)
````

Для всех признаков, кроме пола клиента и типа телефонной линии нулевая гипотеза H0 оказалась отвергнута и принята альтернативная гипотеза H1: признак и целевой признак не являются независимыми. Это согласуется с предположениями графического анализа. Однако стоит заметить, что для `LandlineType` P-значение=0.01034 очень близко к выбраному значению статистической значимости в 1%. Если бы мы выбрали значение в 5%, нулевая гипотеза о независимости признаков была бы опровергнута.

Напишем функцию предобработки данных `preprocess`, которая принимает таблицу с признаками для обработки, дату актуальности данных и список дополнительных признаков, которые надо отбросить (например, можно попробовать отбросить `gender`, который, как мы выяснили во время исследовательского анализа, скорее всего не сильно влияет на вероятность ухода клиента). Функция выполняет все шаги, содержащиеся в разделе «Предобработка данных», плюс отбрасывает дополнительные признаки по выбору. Функция возвращает трансформированную копию данных.

````python
def preprocess(raw_data, ACTUAL_DATE=ACTUAL_DATE, del_columns = []):
    
    data = raw_data.copy()
    
    # пропуски - интернет
    data['InternetService'] = data['InternetService'].fillna(value='No Internet')
    internet_addons = [
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies'
    ]
    data[internet_addons] = data[internet_addons].fillna(value='No')
    
    # пропуски - телефон
    def landline_type(value):
        if value == 'Yes':
            return 'Multiple lines'
        elif value == 'No':
            return 'Single line'
        else:        
            return 'No line'

    data['LandlineType'] = data['MultipleLines'].apply(landline_type)
    data = data.drop('MultipleLines', axis=1)
    
    # длительность договора
    data['BeginDate'] = pd.to_datetime(data['BeginDate'], format='%Y-%m-%d')
    data['EndDate'] = pd.to_datetime(
        data['EndDate'].where(data['EndDate'] != 'No', ACTUAL_DATE),
        format='%Y-%m-%d'
    )
    data['elapsed_months'] = (
        (data['EndDate'].dt.year - data['BeginDate'].dt.year) * 12 + 
        (data['EndDate'].dt.month - data['BeginDate'].dt.month)
    )
    data = data.drop(['BeginDate', 'EndDate'], axis=1)
    
    # типы данных
    data['TotalCharges'] = pd.to_numeric(data['TotalCharges'])

    binary_columns = [col for col in data.columns 
                      if sorted(data[col].unique().tolist()) == ['No', 'Yes']]
    for column in binary_columns:
        data[column] = data[column].apply(lambda x: 1 if x == 'Yes' else 0)  
        
    # интернет-услуги
    data['internet_addons'] = data[internet_addons].sum(axis=1) 
    
    int_cols = [col for col in data.columns 
                if (data[col].dtype == 'int64')]
    data[int_cols] = data[int_cols].astype('int8')
    сat_cols = [col for col in data.columns 
                if data[col].dtype == 'object']
    data[сat_cols] = data[сat_cols].astype('category')
    
    data = data.drop(labels=del_columns, axis=1)
    
    return data
````

С помощью трансформеров `sklearn` стоит также произвести два дополнительных преобразования:

* Масштабировать численные признаки

* Закодировать с помощью One-Hot-Encoding оставшиеся незакодированными категориальные признаки.

Запишем в списки названия численных и категориальных признаков для трансформеров.

````python
numerical_columns = ['MonthlyCharges', 'TotalCharges', 'elapsed_months', 'internet_addons']
categorical_columns = ['Type', 'PaymentMethod', 'InternetService', 'gender', 'LandlineType']
````

`CatBoost` умеет работать с категориальными признаками напрямую, без кодирования, и как правило показвает лучшие результаты при таком использовании. Также было бы интересно попробовать исключать некоторые признаки.

Чтобы не дублировать код создания пайплайна предобработки, напишем функцию `make_preprocessor`. Она принимает аргументы: список признаков для удаления и использовать или нет One-Hot-Encoding. Функция возвращает пайплайн предобработки из следующих щагов:

* `FunctionTransformer` применяет функцию `preprocess`, которой также передает список колонок на удаление.

* `ColumnTransformer` применяет отдельно следующие трансформации:

    * `StandardScaler` для численных признаков
    
    * `OneHotEncoder` для категориальных, если указан аргумент `ohe=True`
    
Списки признаков для трансформеров чистятся от признаков на удаление.

````python
def make_preprocessor(del_columns=[], ohe=True):
    
    def clear_names(columns_list, del_columns=del_columns):
        return [x for x in columns_list if x not in del_columns]

    preprocessor = Pipeline(steps=[
        ('FunctionTransformer', FunctionTransformer(func=preprocess, kw_args={'del_columns': del_columns})),
        ('ColumnTransformer', ColumnTransformer(
                transformers=[
                    ('StandardScaler', StandardScaler(), clear_names(numerical_columns)),
                    ('OneHotEncoder', 
                         OneHotEncoder(handle_unknown='ignore') if ohe else 'passthrough',
                         clear_names(categorical_columns)
                    )
                ], remainder='passthrough'
            )
        )
    ])
    
    return preprocessor
````

Напишем функцию кросс-валидации пайплайна `roc_auc_cv`. Функция принимает пайплайн, признаки и целевой признак, производит кросс-валидацию на 4 складках `StratifiedKFold` с помощью `sklearn.model_selection.cross_val_score`. Функция возвращает **отрицательное** значение средней метрики ROC_AUC для 4 складок. Отрицательное значение нужно, чтобы превратить ROC_AUC в метрику ошибки, которую мы будем минимизировать во время подбора гиперпараметров моделей.

````python
def roc_auc_cv(pipeline, X_train, y_train):
    skf = StratifiedKFold(n_splits=4, shuffle=True, random_state=1) 
    score = cross_val_score(estimator=pipeline, X=X_train, y=y_train, 
                            scoring='roc_auc', cv=skf, n_jobs=-1)
    return -score.mean()
````

Для автоматизации создания пайплайнов с моделью и их кросс-валидации напишем функцию `pipeline_cv`. Она принимает аргументы:

* `model` - модель машинного обучения. Для `CatBoostClassifier` функция создаст список с индексами категориальных признаков в зависимости от выбранного пайплайна предобработки.
* `preprocessor` - пайплайн предобработки. 
* `params` - словарь с параметрами для модели.
* `X_train`, `y_train` - обучающая выборка для кросс-валидации.
* `random_state` - сид генератора случайных чисел для модели.
* `cv` - кросс-валидация. Функция выведет на экран ROC_AUC пайплайна на кросс-валидаци пр значении `True` или вернет готовый пайплайн при значении `False`.
* `show` - показывать ли диагамму модели.

````python
def pipeline_cv(model, preprocessor, params={}, X_train=X_train, y_train=y_train, random_state=123, cv=True, show=True):
    
    if model == CatBoostClassifier:
        transformed = preprocessor.fit_transform(X_train)[0].tolist()
        cat_features = [transformed.index(x) for x in transformed if isinstance(x, str)]
        model = model(**params, random_state=random_state, verbose=False, cat_features=cat_features)
    else:
        model = model(**params, random_state=random_state)
    
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', model)
    ])
    
    if show == True:
        display(pipeline)
    
    if cv == True:
        print('ROC_AUC пайплайна:', abs(roc_auc_cv(pipeline, X_train, y_train)))
    else:
        return pipeline
````

`TotalCharges` и `elapsed_months` - сильная корреляция между этими признаками вполне объяснима: чем дольше клиент пользуется услугами, тем больше он заплатил всего. Попробуем отбросить один из этих признаков во время обучения моделей: это может улучшить их качество.

Для категориальных данных узнаем зависимость признаков, рассчитав критерий Хи-квадрат межу каждой уникальной парой признаков и проверив гипотезы о их зависимости.

````python
def chi_2_categories(del_columns):
    
    passing_columns = [
        item for item in FunctionTransformer(func=preprocess).fit_transform(X_train).columns
        if item not in categorical_columns + numerical_columns
    ]
    columns = categorical_columns + passing_columns
    columns = [x for x in columns if x not in del_columns]

    processed_features = pd.DataFrame(
        data=make_preprocessor(del_columns=del_columns, ohe=False).fit_transform(X_train)[:,4:],
        columns=columns
    )
    
    result = pd.DataFrame()
    idx=0
    significance_level = 0.01
    for feature_1, feature_2 in combinations(columns, 2):
        chisqt = pd.crosstab(processed_features[feature_1], processed_features[feature_2], margins=True)
        value = chisqt[:-1].values
        p, dof = st.chi2_contingency(value)[1:3]
        result.loc[idx, 'Признак_1'] = feature_1
        result.loc[idx, 'Признак_2'] = feature_2
        result.loc[idx, 'Степени свободы'] = dof
        result.loc[idx, 'P-значение'] = p
        if p <= significance_level: 
            result.loc[idx, 'Принятая гипотеза'] = 'H1: признаки не являются независимыми'
        else: 
            result.loc[idx, 'Принятая гипотеза'] = 'H0: признаки независимы'

        result = result.sort_values(by='P-значение')
        idx += 1

    pd.options.display.max_rows = 110
    dependant = result[result['Принятая гипотеза'] == "H1: признаки не являются независимыми"]
    del_recomend = dependant['Признак_2'].unique()
    print('Рекомендовано к удалению:', del_recomend)
    display(result)
````

### Функция подбора гиперпараметров.

Гиперпараметры моделей мы будем подбирать с помощью библиотеки `hyperopt`. Здесь можно реализовать байесовскую оптимизацию параметров. В отличие от случайного поиска или поиска по решетке, данный метод учитывает результаты предыдущих попыток, и, собирая таким образом информацию о функции, пытается сбалансировать зондирование (гиперпараметры, для которых изменение наименее достоверно известно) и использование (гиперпараметры, которые, как ожидается, наиболее близки к оптимуму). На практике это дает лучшие результаты с меньшими вычислениями.

Напишем функцию `hyperopt_search`, принимающую аргументы:
* `space` - пространство гиперпараметров для поиска 
* `model`- модель машинного обучения
* `preprocessor` - пайплайн предобработки
* `max_evals` - количество попыток вычисления значения функции

`hyperopt_search`:

* Создает функцию от гиперпараметров `model_cv`, минимум которой мы будем искать. Функция устроена сходным образом с `pipeline_cv`: строит пайплайн из выбранных модели и предобработчика, получает от `hyperopt.fmin` значения параметров, возвращает отрицательное среднее значение ROC_AUC на кросс-валидации. 

* Производит поиск минимума функции от параметров, используя `hyperopt.fmin`. Выберем алгоритм выбора гиперпараметров из пространства поиска `hyperopt.tpe.suggest` - это байесовская оптимизация.

* Выводит на экран процесс поиска и лучшую комбинацию гиперпараметров.

````python
def hyperopt_search(space, model, preprocessor, max_evals=50):
    
    def model_cv(params, model=model, preprocessor=preprocessor, X_train=X_train, y_train=y_train, random_state=123):
        if model == CatBoostClassifier:
            transformed = preprocessor.fit_transform(X_train)[0].tolist()
            cat_features = [transformed.index(x) for x in transformed if isinstance(x, str)]
            model = model(**params, random_state=random_state, cat_features=cat_features)
        else:
            model = model(**params, random_state=random_state)    
        pipeline = Pipeline(steps=[
            ('preprocessor', preprocessor),
            ('model', model)
        ])

        return roc_auc_cv(pipeline, X_train, y_train)
    
    trials = Trials()

    best_hyperparams=fmin(
        fn=model_cv, 
        space=space, 
        algo=tpe.suggest, 
        max_evals=max_evals, 
        trials=trials, 
        rstate=np.random.RandomState(123) 
    )

    print('\nЛучшие гиперпараметры:\n', space_eval(space, trials.argmin))
    print('\nROC_AUC:', abs(min(trials.losses())))
````

#### StackingClassifier

Попробуем построить ансамбль из наших моделей методом суммарного обобщения - `sklearn.ensemble.StackingClassifier`. Принцип состоит в том, чтобы сложить предсказания разных моделей и использовать их как входные данные для окончательной модели.

Для начала соберем пайплайны для наших моделей с лучшими параметрами и выбором предобработки.

````python
logistic_pipeline = pipeline_cv(
    LogisticRegression,
    make_preprocessor(del_columns=['gender'], ohe=True),
    params={'C': 21.421522042477168, 'penalty': 'l2', 'solver': 'saga', 'max_iter': 300},
    cv=False, show=False
)

forest_pipeline = pipeline_cv(
    RandomForestClassifier,
    make_preprocessor(del_columns=['gender'], ohe=True),
    params={'criterion': 'entropy', 'max_depth': 7, 'n_estimators': 506},
    cv=False, show=False
)

catboost_pipeline = pipeline_cv(
    CatBoostClassifier,
    make_preprocessor(del_columns=['TotalCharges'], ohe=False),
    params={'depth': 4, 'iterations': 194, 'l2_leaf_reg': 3, 'learning_rate': 0.08369509186988548},
    cv=False, show=False
)

estimators = [
    ('LogisticRegression', logistic_pipeline),
    ('RandomForestClassifier', forest_pipeline),
    ('CatBoostClassifier', catboost_pipeline),
]
````

Затем соберем их в ансамбль, добавив логистическую регрессию как финальную модель.

## Тестирование моделей

Соберем пайплайны и их названия в словарь.

````python
pipelines = {
    'Логистическая регрессия': logistic_pipeline,
    'Случайный лес': forest_pipeline,
    'Градиентный бустинг': catboost_pipeline,
    'Ансамбль моделей': stacking_pipeline
}
````

Проитерируемя по словарю, обучим каждый пайплайн на обучающей выборке, сделаем предсказания на тестовой. Соберем в таблицу для каждого пайплайна метрики ROC_AUC, Precision, Recall, F1-метрику, Accuracy, а также время обучения и время предсказания.

````python
result = pd.DataFrame()
for pipeline_name, pipeline in pipelines.items():
    
    fit_start = time.time()
    pipeline.fit(X_train, y_train)
    fit_elaplsed_time = time.time() - fit_start
    
    predict_start = time.time()
    predicted = pipeline.predict(X_test)
    predict_elaplsed_time = time.time() - predict_start
    
    proba = pipeline.predict_proba(X_test)[:,1]
    
    result.loc[pipeline_name, 'ROC_AUC'] = roc_auc_score(y_test, proba)
    result.loc[pipeline_name, 'Precision'] = precision_score(y_test, predicted)
    result.loc[pipeline_name, 'Recall'] = recall_score(y_test, predicted)
    result.loc[pipeline_name, 'F1-метрика'] = f1_score(y_test, predicted)
    result.loc[pipeline_name, 'Accuracy'] = accuracy_score(y_test, predicted)
    result.loc[pipeline_name, 'Время обучения'] = fit_elaplsed_time
    result.loc[pipeline_name, 'Время предсказания'] = predict_elaplsed_time
    
display(result)
````

* Все модели обучены на обучающей выборке, для них сделаны предсказания на тестовой. Для моделей рассчитаны: ROC_AUC, Precision, Recall, F1-метрика, Accuracy, а также время обучения и время предсказания.

* **Лучший результат на тестовой выборке у модлеи градиентного бустинга: ROC_AUC=0.8571**. Ансамбль показал практически такой же результат, однако у него ниже метрики Recall, F1-метрику, Accuracy, плюс ощутимо большее время работы (что в принипе не особо критично для данного задания). На тестовой выборке градиентый бустинг показал себя сильно лучше, чем на кросс-валидации. Скорее всего это говорит о том, что CatBoost более устойчив к переобучению, хотя, возможно это объясняется особенностями выборки.

* Для лучшей модели построены матрица ошибок, ROC кривая и график FPR и TPR модели для разных порогов предсказания.

* Из интерпретации работы лучшей модели наблюдаем, что с большим отрывом самые важные признаки - elapsed_months, Type, MonthlyCharges. Компании стоит сосредоточить внимание на недвано пришедших клентах, клиентах с ежемесячным типом оплаты, клиентах с большой ежемесячной суммой оплаты.
